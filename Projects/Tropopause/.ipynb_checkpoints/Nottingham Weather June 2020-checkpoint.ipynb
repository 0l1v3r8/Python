```python
#Surface Weather Data from June 26-28 2020
surface_data = pd.read_csv("NottinghamSurfaceWeatherData23to28-6-2020.csv")
surface_data.columns = surface_data.iloc[0,:]
surface_data = surface_data.iloc[1:,:]
surface_data
```

```python
surface_data["Date_Time"] = surface_data["Date"]+ " " + surface_data["Time"]
surface_data["Date_Time"] = pd.to_datetime(surface_data["Date_Time"],format="%d/%m/%Y %H:%M:%S")
surface_data
```

```python
def max_min(column , a, belowZero=True):
    maxi = ((column.max()//a)*a) + a
    mini = ((column.min()//a)*a) - a
    if (not belowZero) and (mini<0):
        mini = 0
    rang = np.arange(mini,maxi+a,a)
    return [[maxi,mini] , rang]
```

```python
surface_data["Temperature(C)"] = surface_data["Temperature(C)"].apply(float)
label = []
for i in range(len(surface_data)):
    j = i
    if (j % 12 == 0):
        day = surface_data["Date_Time"].iloc[i].day
        month = surface_data["Date_Time"].iloc[i].month
        hour = surface_data["Date_Time"].iloc[i].hour
        label.append(f"{day}/{month}\n{hour}:00")

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)
idx = []
for i in range(1,len(surface_data["Date_Time"]),12):
    idx.append(i)

plt.plot(surface_data["Date_Time"],surface_data["Temperature(C)"],".-", color="navy")
plt.xticks(surface_data["Date_Time"][idx],label)
plt.yticks(max_min(surface_data["Temperature(C)"],1)[1])
plt.title("Change in Surface Dry Bulb Temperature(C) with Time \n23-28/6/2020",weight="bold")
plt.ylabel("Temperature(C)", weight="bold")
plt.xlabel("Date and Time", weight="bold")
plt.grid()
plt.show()
```

```python
surface_data["Dew Point(C)"] = surface_data["Dew Point(C)"].apply(float)
label = []
for i in range(len(surface_data)):
    j = i
    if (j % 12 == 0):
        day = surface_data["Date_Time"].iloc[i].day
        month = surface_data["Date_Time"].iloc[i].month
        hour = surface_data["Date_Time"].iloc[i].hour
        label.append(f"{day}/{month}\n{hour}:00")

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)
idx = []
for i in range(1,len(surface_data["Date_Time"]),12):
    idx.append(i)

plt.plot(surface_data["Date_Time"],surface_data["Dew Point(C)"],".-", color="navy")
plt.xticks(surface_data["Date_Time"][idx],label)
plt.yticks(max_min(surface_data["Dew Point(C)"],1)[1])
plt.title("Change in Surface Dew Point Temperature with Time \n23-28/6/2020",weight="bold")
plt.ylabel("Dew Point Temperature (C)", weight="bold")
plt.xlabel("Date and Time", weight="bold")
plt.grid()
plt.show()
```

```python
def potential_temp(T,p):
    #T in Celcius, p in hPa, returns pot_temp in Celcius
    p0 = 1000
    Tk= T+273.15
    k = 2/7 #poisson constant
    pot_tempk = Tk*((p0/p)**k)
    pot_temp = pot_tempk - 273.15
    return pot_temp
```

```python
surface_data["Pressure (mb)"] = surface_data["Pressure (mb)"].apply(float)
surface_data["Potential Temperature(C)"] = potential_temp(surface_data["Temperature(C)"],surface_data["Pressure (mb)"])
surface_data[["Potential Temperature(C)"]]
```

```python
label = []
for i in range(len(surface_data)):
    j = i
    if (j % 12 == 0):
        day = surface_data["Date_Time"].iloc[i].day
        month = surface_data["Date_Time"].iloc[i].month
        hour = surface_data["Date_Time"].iloc[i].hour
        label.append(f"{day}/{month}\n{hour}:00")

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)
idx = []
for i in range(1,len(surface_data["Date_Time"]),12):
    idx.append(i)

plt.plot(surface_data["Date_Time"],surface_data["Potential Temperature(C)"],".-", color="navy")
plt.xticks(surface_data["Date_Time"][idx],label)
plt.yticks(max_min(surface_data["Potential Temperature(C)"],1)[1])
plt.title("Change in Surface Potential Temperature(C) with Time \n23-28/6/2020",weight="bold")
plt.ylabel("Potential Temperature(C)", weight="bold")
plt.xlabel("Date and Time", weight="bold")
plt.grid()
plt.show()
```
