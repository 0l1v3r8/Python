```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


def select_sample(mean,std,n):
    sample = np.random.normal(mean,std,n)
    return sample


rng = np.arange(1,10,1)
rng10 = 10**rng
stdevs = np.array([])

for i in rng10:
    sample = select_sample(50,3,i)
    std = sample.std() 
    stdevs = np.append(std ,stdevs)
    plt.hist(sample,ec="w")
    plt.show()
    print("Standard Deviation:",std)
    
```

```python
plt.plot(rng,stdevs, type="line")
```
